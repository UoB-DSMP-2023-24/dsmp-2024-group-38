## Balance Imputation Code
 

```python
bank = df.copy()

for i in range(len(bank)):
    
    if pd.isnull(bank.loc[i,'Balance']):
        bank.loc[i,'Balance'] = bank.loc[i-1,'Balance'] + bank.loc[i,'Amount']
```
