# EDA

```python
# Clean and preprocess data
data['DateTime'] = pd.to_datetime(data['DateTime'])
data.set_index('DateTime', inplace=True)

# Filter for negative transactions
negative_transactions = data[data['Amount'] < 0]

# Group by day of the week and aggregate
daily_negatives_per_weekday = negative_transactions.groupby(negative_transactions.index.day_name()).agg({'Amount': ['sum', 'count']})

# Order days of the week correctly
days_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
daily_negatives_per_weekday = daily_negatives_per_weekday.reindex(days_order)

# Plotting
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14, 6))
fig.suptitle('Summary of Negative Transactions for Each Weekday in 2023', fontsize=16)

# Total amount per weekday
daily_negatives_per_weekday['Amount']['sum'].plot(ax=axes[0], kind='bar', title='Total Amount per Weekday (Negative Transactions)')
axes[0].set_ylabel('Total Amount ($)')
axes[0].set_xlabel('Day of the Week')
# Transaction count per weekday
daily_negatives_per_weekday['Amount']['count'].plot(ax=axes[1], kind='bar', title='Transaction Count per Weekday (Negative Transactions)')
axes[1].set_ylabel('Transaction Count')
axes[1].set_xlabel('Day of the Week')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
data = data[data['Amount'] < 0]
data['DateTime'] = pd.to_datetime(data['DateTime'])
data.set_index('DateTime', inplace=True)
data['Hour'] = data.index.hour

# Group by hour and aggregate
hourly_transactions = data.groupby('Hour').agg({'Amount': ['sum', 'count']})

# Plotting
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14, 6))
fig.suptitle('Hourly Summary of Transactions for the Entire Dataset', fontsize=16)

# Total amount per hour
hourly_transactions['Amount']['sum'].plot(ax=axes[0], kind='bar', title='Total Transaction Amount per Hour')
axes[0].set_ylabel('Total Amount ($)')
axes[0].set_xlabel('Hour of the Day')
# Transaction count per hour
hourly_transactions['Amount']['count'].plot(ax=axes[1], kind='bar', title='Transaction Count per Hour')
axes[1].set_ylabel('Transaction Count')
axes[1].set_xlabel('Hour of the Day')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
# Grouping the data by 'Account No' and 'Third Party Name' to calculate total transaction frequency and total amount
df2 = df[df['Amount'] < 0]
grouped_df = df2.groupby(['Account No', 'Third Party Name']).agg(
    Total_Transactions=('Amount', 'count'),
    Total_Amount=('Amount', 'sum')
).reset_index()
grouped_df
```

```python
# Pivot the table to get 'Third Party Name' as columns with total transactions and amounts for each account
pivot_df = pd.pivot_table(grouped_df, values=['Total_Transactions', 'Total_Amount'], 
                          index='Account No', columns='Third Party Name', fill_value=0)

# Flattening the multi-level column names
pivot_df.columns = [' '.join(col).strip() for col in pivot_df.columns.values]

pivot_df.reset_index(inplace=True)
pivot_df
```
