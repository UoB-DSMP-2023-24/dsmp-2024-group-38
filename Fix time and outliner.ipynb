# Time and Amount

```python
data['DateTime'] = pd.to_datetime(data['DateTime'])
data['Date'] = data['DateTime'].dt.date
data['Time'] = data['DateTime'].dt.time
data
```

```python
Fixed_time = ~data['Time'].isin([pd.to_datetime('00:00').time(), pd.to_datetime('23:59').time()])
data2 = data[Fixed_time]
data2 = data2[data2['Amount'] < 0]
data2
```

#### Only use Amount < 0 and not happen in fixed time

```python
df = data2.copy()
```

```python
df['Time'] = pd.to_datetime(df['Time'], format='%H:%M:%S').dt.time
df['Time_seconds'] = df['Time'].apply(lambda t: t.hour * 3600 + t.minute * 60 + t.second)

fig, ax = plt.subplots(1, 2, figsize=(18,4))

amount_val = df['Amount'].values
time_val = df['Time_seconds'].values

sns.distplot(amount_val, ax=ax[0], color='r')
ax[0].set_title('Distribution of Transaction Amount', fontsize=14)
ax[0].set_xlim([min(amount_val), max(amount_val)])

sns.distplot(time_val, ax=ax[1], color='b')
ax[1].set_title('Distribution of Transaction Time', fontsize=14)
ax[1].set_xlim([min(time_val), max(time_val)])

plt.show()
```
