## separate

```python
df
```

```python
# Convert Time column to seconds
df['Time'] = pd.to_datetime(df['Time'], format='%H:%M:%S').dt.time
df['Time_seconds'] = df['Time'].apply(lambda t: t.hour * 3600 + t.minute * 60 + t.second)

# Setting the style and color palette
sns.set(style="whitegrid")
color_palette = sns.color_palette("viridis", as_cmap=False)
lloyds_green = "#007F3B"


# Plot Distribution of Transaction Time with improved aesthetics
fig, ax = plt.subplots(figsize=(12, 6))
sns.histplot(df['Time_seconds'], ax=ax, color=lloyds_green, kde=True, binwidth=3600)
ax.set_title('Distribution of Transaction Time', fontsize=16)
ax.set_xlabel('Time of Day', fontsize=12)
ax.set_ylabel('Density', fontsize=12)

# Set explicit range from 00:00 to 24:00
ax.set_xlim([0, 24 * 3600])

# Adjust x-ticks to hourly intervals and format them into readable hours
hours = [3600 * x for x in range(0, 25, 3)]
hour_labels = [f"{x//3600}:00" for x in hours]
ax.set_xticks(hours)
ax.set_xticklabels(hour_labels)
plt.show()
```

```python
from scipy.signal import savgol_filter

# Filter outliers using the IQR method
Q1 = df['Amount'].quantile(0.25)
Q3 = df['Amount'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# remove outliers and change abs of Amount value
filtered_df = df[(df['Amount'] >= lower_bound) & (df['Amount'] <= upper_bound)]
filtered_df['Amount'] = filtered_df['Amount'].abs()

# Calculate average transaction amount by time
average_amount_by_time = filtered_df.groupby('Time_seconds')['Amount'].mean().reset_index()
average_amount_by_time['Smoothed'] = savgol_filter(average_amount_by_time['Amount'], window_length=101, polyorder=2)


# Plot the average transaction amount by time
fig, ax = plt.subplots(figsize=(14, 6))
sns.lineplot(x='Time_seconds', y='Smoothed', data=average_amount_by_time, ax=ax, color="#007F3B")
ax.set_title('Average Transaction Amount by Transaction Time', fontsize=16)
ax.set_xlabel('Time of Day', fontsize=14)
ax.set_ylabel('Average Amount ($)', fontsize=14)

ax.set_xlim([0, 24 * 3600])
hours = [3600 * x for x in range(0, 25, 3)]
hour_labels = [f"{x//3600}:00" for x in hours]
ax.set_xticks(hours)
ax.set_xticklabels(hour_labels)

plt.show()
```

## Weekly and monthly

```python
df = data3.copy()
df['Amount'] = df['Amount'].abs()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.month

# Group by 'Month' and sum the 'Amount' column
monthly_totals = df.groupby('Month')['Amount'].sum().reset_index()

# Plot the bar chart
plt.figure(figsize=(12, 6))
sns.barplot(data=monthly_totals, x='Month', y='Amount')
plt.xlabel('Month')
plt.ylabel('Total Transaction Amount (£)')
plt.title('Total Amount by Month')
plt.show()
```

```python
colors = ["#A83E3E", "#CF7D72", "#E8D4C6", "#D6AB96", "#DBB18C", "#B8CFDC", "#7399AD", "#374C72"]

# Extract the day of the week from the 'Date' column
df['Date'] = pd.to_datetime(df['Date'])
df['Weekday'] = df['Date'].dt.day_name()

# Group by 'Weekday' and sum the 'Amount' column
weekday_totals = df.groupby('Weekday')['Amount'].sum().reset_index()

# Ensure the correct order of days
weekday_order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
weekday_totals['Weekday'] = pd.Categorical(weekday_totals['Weekday'], categories=weekday_order, ordered=True)
weekday_totals = weekday_totals.sort_values('Weekday')


plt.figure(figsize=(12, 6))
sns.barplot(data=weekday_totals, x='Weekday', y='Amount', palette=colors)
plt.xlabel('Weekday')
plt.ylabel('Total Amount (£)')
plt.title('Total Transaction Amount by Weekday')
plt.show()
```

# outliner

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from pyod.models.ecod import ECOD
from pyod.utils.example import visualize

X = data[['Time_seconds', 'Amount']].values

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

clf_name = 'ECOD'
clf = ECOD()
clf.fit(X_scaled)

y_pred = clf.labels_  # binary labels (0: inliers, 1: outliers)
scores = clf.decision_scores_  # raw outlier scores

print("Detected outliers (indices):", np.where(y_pred == 1)[0])
print("Outlier scores:", scores[y_pred == 1])

if X_scaled.shape[1] == 2:
    visualize(clf_name, X_scaled, y_pred, X_scaled, y_pred, y_pred, y_pred, show_figure=True, save_figure=False)
else:
    print("Visualisation requires exactly 2 features.")
```

```python
df = data3.copy()
df['Amount'] = df['Amount'].abs()
```

```python
df['Hour'] = pd.to_datetime(df['DateTime']).dt.hour
hourly_threshold = df.groupby('Hour')['Amount'].quantile(0.995)

# Identify outliers based on the threshold
df['Upper_Threshold'] = df['Hour'].map(hourly_threshold)
df['Is_Outlier'] = df['Amount'] > df['Upper_Threshold']

# Plot the distribution and highlight outliers
plt.figure(figsize=(12, 8))
sns.scatterplot(data=df, x='Hour', y='Amount', color='orange', alpha=0.7, label='Regular Transactions')
sns.scatterplot(data=df[df['Is_Outlier']], x='Hour', y='Amount', color='blue', label='Outliers')
plt.plot(hourly_threshold.index, hourly_threshold.values, linestyle='--', color='green', linewidth=2, label='Upper Threshold')
plt.xlabel('Hour of Day')
plt.ylabel('Amount')
plt.title('Transaction Amounts by Hour')
plt.legend()
plt.grid(True)
plt.show()
```

## separate

```python
account_counts = df['Account No'].value_counts()

plt.figure(figsize=(10, 6))
plt.hist(account_counts, bins=10, color='skyblue', edgecolor='black')
plt.xlabel('Transaction Count')
plt.ylabel('Number of Accounts')
plt.title('Transaction Count Distribution by Account')
plt.grid(True)
plt.show()
```

```python
amount_counts = df.groupby('Account No')['Amount'].sum()

plt.figure(figsize=(10, 6))
plt.hist(amount_counts, bins=10, color='skyblue', edgecolor='black')
plt.xlabel('Total Amount (£)')
plt.ylabel('Number of Accounts')
plt.title('Transaction Amount Distribution by Account')
plt.grid(True)
plt.show()
```

## combine

```python
combined_df = pd.merge(account_counts, amount_counts, on='Account No')
combined_df
```

```python
# Adding mean lines to the histograms

mean_count = int(combined_df['count'].mean())
mean_amount = int(combined_df['Amount'].mean())

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14, 6))

# Histogram for 'count' with mean line
combined_df['count'].plot(kind='hist', bins=30, ax=axes[0], color='skyblue')
axes[0].axvline(mean_count, color='black', linestyle='dashed', linewidth=1)
axes[0].text(mean_count + 10, 20, f'Mean: {mean_count}', color='black')
axes[0].set_title('Distribution of Transaction Counts')
axes[0].set_xlabel('Count')
axes[0].set_ylabel('Number of Account')

# Histogram for 'Amount' with mean line
combined_df['Amount'].plot(kind='hist', bins=30, ax=axes[1], color='salmon')
axes[1].axvline(mean_amount, color='black', linestyle='dashed', linewidth=1)
axes[1].text(mean_amount + 1000, 20, f'Mean: {mean_amount}', color='black')
axes[1].set_title('Distribution of Transaction Amounts')
axes[1].set_xlabel('Amount ($)')
axes[1].set_ylabel('Number of Account')

plt.tight_layout()
plt.savefig('C:/Users/Demi/Desktop/plot1.png')
plt.show()
```
